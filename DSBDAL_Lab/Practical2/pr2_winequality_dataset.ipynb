```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from scipy import stats

df=pd.read_csv(file_path)
print(df.shape)
print(df.head())

```

```python
data=df.drop('quality',axis=1)
```

```python
print(data.shape)
data.head()
```

```python
# Outlier detection
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data)
plt.title("Boxplot of Wine Features")
plt.xticks(rotation=45)
plt.show()
```

```python
# Z-Score method

zscores=np.abs(stats.zscore(data))
print("Z- Scores: ")
print(zscores)

outliers=np.where(zscores>3)
print("\nOutliers:\n")
print(list(zip(outliers[0],outliers[1])))
print("\nTotal outliers:", np.sum(zscores > 3))
```

```python
print("Outlier location(row,col):\n")
print(list(zip(outliers[0][:10],outliers[1][:10])))
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data)
plt.xticks(rotation=45)
plt.title("Before Removing Outliers")
plt.show()
```

```python
mask=(zscores<3).all(axis=1)

clean_data=data[mask]
print("Original data shape:", data.shape)
print("Cleaned data shape:", clean_data.shape)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=clean_data)
plt.xticks(rotation=45)
plt.title("After Removing Outliers (Z-score < 3)")
plt.show()
```

```python
for col in data.columns:
    plt.figure(figsize=(4,3))
    sns.histplot(data[col], kde=True)
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = data.columns
n = len(cols)

rows = (n + 2) // 3   # 3 plots per row
plt.figure(figsize=(15, 4 * rows))

for i, col in enumerate(cols, 1):
    plt.subplot(rows, 3, i)
    sns.histplot(data[col], kde=True)
    plt.title(f"Skewness = {data[col].skew():.2f}")
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
print("Before:", data['residual sugar'].skew())
sns.histplot(data['residual sugar'], kde=True)
```

```python
data['residual sugar log'] = np.log1p(data['residual sugar'])
```

```python
print("After:", data['residual sugar log'].skew())
sns.histplot(data['residual sugar log'], kde=True)
```
