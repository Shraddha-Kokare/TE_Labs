```python
df=pd.read_csv("Salary_Data.csv")
```

```python
df
```

```python
plt.scatter(df["YearsExperience"],df["Salary"])
```

```python
bins=[0,2,5,10,df["Salary"].max()]
categories=["Fresher","Junior","MidLevel","Senior"]
```

```python
df["ExperienceLevel"]=pd.cut(df["YearsExperience"],bins=bins,labels=categories,include_lowest=True)
```

```python
df
```

```python
df.groupby("ExperienceLevel")["Salary"].describe()
```

```python
df.boxplot(column="Salary",by="ExperienceLevel")
```

```python
plt.hist(df["Salary"])
plt.xlabel("Salary")
plt.ylabel("Frequency")
```
